```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, KFold
y = df["is_canceled"]
x = df.drop("is_canceled", axis=1)
train_raw, test_raw, target, target_test = train_test_split(x,y, test_size=0.2, stratify= y, random_state=0)
```

```python
categorical_features1 = x.select_dtypes(include="object").columns
integer_features1 = x.select_dtypes(exclude="object").columns
full_pipeline = ColumnTransformer([
        ("num", StandardScaler(), integer_features1),
        ("cat", OneHotEncoder(), categorical_features1),
    ])
pipeline = Pipeline([
    ('scaler', StandardScaler())
])
train = full_pipeline.fit_transform(train_raw)
test = full_pipeline.fit_transform(test_raw)
```

```python
hotel_prepared
```

```python
from sklearn.cluster import KMeans

ks = list(range(2,30))
inertia = []
for k in ks:
    kmeans = KMeans(n_clusters = k, init = 'k-means++', random_state = 0)
    kmeans.fit(train)
    # inertia method returns wcss for that model
    inertia.append(kmeans.inertia_)
    print(f"Inertia for K = {k}: {kmeans.inertia_}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(ks, inertia,marker='o',color='red')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
